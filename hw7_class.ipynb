```python
from matplotlib import pyplot as plt
import numpy as np
import tkinter as tk
from tkinter import filedialog
```

```python
class SelectPlot(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self.createWidgets()
        self.grid()
    def plot(self):
        self.filename=filedialog.askopenfilename(title = "select")
        lines=np.loadtxt(self.filename).transpose()
        plt.plot(lines[0],lines[1])
        plt.show()
    def createWidgets(self):
        self.findbutton= tk.Button(self,text='select',command=self.plot)
        self.findbutton.grid() 
app = SelectPlot()
app.title('6666')
app.mainloop()
```
